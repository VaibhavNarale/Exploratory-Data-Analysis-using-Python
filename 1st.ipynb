# Eploratory Data Analysis

# Customer Segment

```python
ax = sns.countplot(x='cus_seg', data=df, color='red')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen = df.groupby(['cus_seg'], as_index=False)['sales'].sum().sort_values(by='sales', ascending=False)
#print(sales_gen)
ax=sns.barplot(y='sales' ,x='cus_seg',data=sales_gen,color='gray')
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d') #fmd='%d' covert mu complex value into int
plt.show()
```

The Corporate segment visits the hotel the most frequently, followed by Individual and Family segments.
Despite fewer visits, Family guests contribute significantly to total sales, nearly matching Corporate guests, suggesting higher spending per visit.
# On Which Segment we need to focus and why 
1) High Spending Potential: Despite having the fewest visits, the Family segment generates almost as much total revenue as Corporate guests. This indicates they tend to spend more per visit.
2) Growth Opportunity: Since Family visits are lower, there's room to attract more families through targeted marketing, family-friendly packages, and special offers. Increasing their visit frequency could significantly boost overall revenue.


# Sales Persons

```python
sales_gen1 = df.groupby(['Sales Person '], as_index=False)['sales'].sum().sort_values(by='sales', ascending=False)
#print(sales_gen1)
ax=sns.barplot(y='sales' ,x='Sales Person ',data=sales_gen1,color='gray')
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
plt.show()
```

```python
comm=df.groupby(['Sales Person '], as_index=False)['comm_pay'].sum()
print(comm)
ax=sns.barplot(y='comm_pay' ,x='Sales Person ',data=comm)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

# Leo generates the highest total sales, followed by Amy and Derek, indicating Leo's strong contribution to overall revenue.
# However, Amy has the highest average commission pay, suggesting she may be closing more high-margin or commission-heavy deals despite lower total sales. 

# Payment Mode

```python
ax=sns.countplot(x='state',data=df,hue='payment_method')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
